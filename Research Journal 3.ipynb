## Describe your results


Understanding why these parameters were decreased:

The Eddington factor affects how quickly supermassive black holes reach their observed masses. If accretion stays efficient despite a large Eddington factor, black holes could grow much faster than previously predicted, which could explain the existence of these observed supermassive black holes at high redshifts [1]. The energy ejected by the supernova feedback is known as wind energy.  By default, 3.6 times x10^15 of the energy being expelled from these winds is ejected.  When energy is reduced, more gas is present in the environment, which promotes black hole growth [2]. These winds carry away energy and momentum, influencing star formation in the host galaxy. When these winds decrease in velocity, less energy is carried away, allowing more gas to remain in the accretion disk. This excess gas fuels further black hole growth, leading to increased in mass. The efficiency tells us how much of the gas turns into energy and how that energy impacts the environment. If the efficiency is lower, less energy is released, meaning more gas stays around for the black hole to consume, allowing it to grow faster.
